```python
import tensorflow as tf
import numpy as np
import cv2
```

```python
# Load your trained model
model = tf.keras.models.load_model(r"C:\Users\manic\OneDrive\Desktop\Cancer detection\Website\lung_cancer_model.h5")

# Load and preprocess the input image
def preprocess_image(image_path):
    img = cv2.imread(image_path)  # Read the image
    img = cv2.resize(img, (224, 224))  # Resize to match model input size
    img = img / 255.0  # Normalize pixel values
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img
```

```python
# Provide the path to an image
image_path = r"C:\Users\manic\OneDrive\Desktop\Cancer detection\lung_colon_image_set\lung_image_sets\lung_scc\lungscc238.jpeg"

# Make a prediction
prediction = model.predict(image_path)
predicted_class = np.argmax(prediction, axis=1)[0]  # Get class index

# Map prediction to class labels
class_labels = {0: "Lung Adenocarcinoma (ACA)", 1: "Normal Lung Tissue", 2: "Lung Squamous Cell Carcinoma (SCC)"}
predicted_label = class_labels[predicted_class]

print(f"Predicted Class: {predicted_label}")
```
